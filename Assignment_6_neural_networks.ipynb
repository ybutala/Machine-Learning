```python
import numpy as np
import pickle
import torch
from torch import nn, from_numpy
from torch.optim import Adam
from sklearn.utils import shuffle
from torch.utils.data import TensorDataset, DataLoader
from torchvision import datasets, transforms
```

```python
X_train,X_test,y_train,y_test = pickle.load(open('/content/drive/My Drive/ML_Fundamentals/nn/fashion_mnist.p','rb'))
```

#Q6.1a

```python
784*128 + 128 + 128*10 + 10
```

```python
X_train = from_numpy(X_train)
X_test = from_numpy(X_test)
y_train = from_numpy(y_train)
y_test = from_numpy(y_test)
```

#Q6.1b

```python
inputlayer = 784
nodes1 = 128
linearlayer = 10

firstlayerp = (inputlayer + 1) * nodes1

secondlayerp = (nodes1 + 1) * linearlayer

totalp = firstlayerp + secondlayerp

model = nn.Sequential(
    nn.Linear(784, 128),
    nn.ReLU(),
    nn.Linear(128, 10)
)
loss_function = nn.CrossEntropyLoss()
optimizer = Adam(model.parameters())
```

```python
n_iter = 150

for t in range(n_iter):
    print('Running iteration ' + str(t))
    y_pred = model(X_train)

    # Forward pass
    print(model(X_train[0]))
    loss = loss_function(y_pred, y_train)

    # Backward pass
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    print('In-sample cross-entropy: ' + str(loss.item()))

y_pred = model(X_test)
print('Out-of-sample accuracy: ' + str((y_pred.argmax(1) == y_test).sum().item() / 10000))
```

#Q6.1d

```python
model = nn.Sequential(
    nn.Linear(784, 128),
    nn.ReLU(),
    nn.Linear(128, 32),
    nn.ReLU(),
    nn.Linear(32, 10)
)
loss_function = nn.CrossEntropyLoss()
optimizer = Adam(model.parameters())

n_iter = 150

for t in range(n_iter):
    print('Running iteration ' + str(t))
    y_pred = model(X_train)

    # Forward pass
    print(model(X_train[0]))
    loss = loss_function(y_pred, y_train)

    # Backward pass
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    print('In-sample cross-entropy: ' + str(loss.item()))

y_pred = model(X_test)
print('Out-of-sample accuracy: ' + str((y_pred.argmax(1) == y_test).sum().item() / 10000))
```

We see the accuracy has increased by approximately 3% for same number of iterations.

#Q6.2

```python
X_train = np.reshape(X_train, (60000,1,28,28))
X_test  = np.reshape(X_test, (10000,1,28,28))

X_train_tensor = X_train
X_test_tensor = X_test
y_train_tensor = y_train
y_test_tensor = y_test

```

#Q6.2a

```python
model = nn.Sequential(nn.Conv2d(1,40,6),
        nn.ReLU(),
        nn.Flatten(),
        nn.Linear(40*23*23,10))
loss_function = nn.CrossEntropyLoss()
optimizer = Adam(model.parameters())

n_epochs = 20
batch_size = 1000
for epoch in range(n_epochs):
    print('Running epoch ' + str(epoch))
    X_train_tensor, y_train_tensor = shuffle(X_train_tensor, y_train_tensor)

    # Ea. cycle is a loop through all the batches:
    for batch in range(int(60000/batch_size)):

        # ForwardProp
        y_pred = model(X_train_tensor[batch*batch_size:(batch+1)*batch_size])
        loss = loss_function(y_pred, y_train_tensor[batch*batch_size:(batch+1)*batch_size])

        # BackwardProp
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    # Printing in-sample & out-of-sample accuracy
    print('In-sample cross-entropy: ' + str(loss.item()))
    y_pred = model(X_test_tensor)
    print('Out-of-sample accuracy: ' + str((y_pred.argmax(1) == y_test_tensor).sum().item() / 10000))
```

There was an improvement of 6 % in the case of convolutional neural networks as compared to neural networks
